```python
def phi1(a, b):
    if a == b:
        return 4
    else:
        return 2
    
def phi2(b, c):
    if b == c:
        return 5
    else:
        return 1
    
def phi3(c, a):
    if c == a:
        return 3
    else:
        return 1
```

```python
def all_comp():
    
    result_dict = {}
    
    for a in [0,1]:
        for b in [0,1]:
            for c in [0,1]:
                fac1= phi1(a,b)
                fac2= phi2(b,c)
                fac3= phi3(c,a)
                combination = fac1 * fac2 * fac3
                
                result_dict[(a,b,c)] = combination
    
    return result_dict
```

```python
joint = all_comp()
print(joint)
```

```python
def partition_function(joint):
    Z = 0
    for value in joint.values():
        Z += value
    return Z
```

```python
def normalize(joint):
    Z = partition_function(joint)
    normalized = {}
    
    for (a, b, c), unnorm_val in joint.items():
        norm_prob = unnorm_val / Z
        normalized[(a, b, c)] = norm_prob
    
    return normalized

def show_normalized_table(joint):
    normalized = normalize(joint)
    
    for (a, b, c), prob in normalized.items():
        phi1_val = phi1(a, b)
        phi2_val = phi2(b, c)
        phi3_val = phi3(c, a)
        factors_str = str(phi1_val) + "." + str(phi2_val) + "." + str(phi3_val)
        print("A" + str(a) + "  B" + str(b) + "  C" + str(c) + "  " + factors_str + "  " + str(round(prob, 6)))

joint = all_comp()
show_normalized_table(joint)
```

```python
def query(a, b, c):
    joint = all_comp()
    Z = partition_function(joint)
    unnorm_val = joint[(a, b, c)]
    norm_prob = unnorm_val / Z
    return norm_prob

print(query(1, 1, 1))
print(query(0, 1, 0))
```

```python
def marginal(variable:str, value: int):
    marginal_prop = 0
    for a in [value] if variable=='A' else [0, 1]:
        for b in [value] if variable=='B' else [0, 1]:
            for c in [value] if variable=='C' else [0, 1]:
                marginal_prop += query(a, b, c)
    
    return marginal_prop

print(marginal('A', 1))
print(marginal('C', 0))
```
