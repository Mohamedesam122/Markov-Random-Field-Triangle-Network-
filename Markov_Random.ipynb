```python
def phi1(a, b):
    if a == b:
        return 4
    else:
        return 2
    
def phi2(b, c):
    if b == c:
        return 5
    else:
        return 1
    
def phi3(c, a):
    if c == a:
        return 3
    else:
        return 1
```

```python
def all_comp():
    
    result_dict = {}
    
    for a in [0,1]:
        for b in [0,1]:
            for c in [0,1]:
                fac1= phi1(a,b)
                fac2= phi2(b,c)
                fac3= phi3(c,a)
                combination = fac1 * fac2 * fac3
                
                result_dict[(a,b,c)] = combination
    
    return result_dict
```

```python
joint = all_comp()
print(joint)
```
